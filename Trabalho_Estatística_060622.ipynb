```python
import pandas as pd
import numpy as np
from scipy.stats import variation 
```

Baixe o arquivo dados do consumidor 2017 disponível aqui e calcule
algumas medidas descritivas.


```python
df = pd.read_csv('dadosgovbr---2017.csv', sep=';', encoding='latin-1')

pd.set_option('display.max_columns', 38)
df.head()
```

```python
df.shape
```

1) Apresente as medidas de posição e de tendência central das
variáveis quantitativas.

```python
df.describe()
```

```python
# Avalição dos consumidores.

media_nota = df['Nota do Consumidor'].mean()
print(f'A média da avaliação dos consumidores é {media_nota: .2f}.')

mediana_nota = df['Nota do Consumidor'].median()
print(f'A mediana da avaliação do consumidores é {mediana_nota}.')

moda = df['Nota do Consumidor'].mode()[0]
print(f'A moda da avalição dos cosnumidores é {moda}.')

desv_padrao = df['Nota do Consumidor'].std()
print(f'O desvio padrão da avaliação dos consumidores é {desv_padrao: .2f}.')

variancia = df['Nota do Consumidor'].var()
print(f'A variância avaliação dos consumidores é: {variancia: .2f}.')

coef_var = (desv_padrao / media_nota) * 100
print(f'O coeficiente de variação da avaliação dos consumidores é:{coef_var: .2f}%.')
```

```python
# Resposta dos consumidores.

media_tempo = df['Tempo Resposta'].mean()
print(f'A média do tempo de resposta é de: {media_tempo: .2f}.')

mediana_nota = df['Tempo Resposta'].median()
print(f'A mediana do tempo de resposta é de: {mediana_nota: .2f}.')

moda = df['Tempo Resposta'].mode()[0]
print(f'A moda do tempo de resposta é de: {moda: .2f}.')

desv_padrao = df['Tempo Resposta'].std()
print(f'O desvio padrão do tempo de resposta é de:{desv_padrao: .2f}.')

variancia = df['Tempo Resposta'].var()
print(f'A variância do tempo de resposta é de: {variancia: .2f}.')

coef_var = (desv_padrao / media_nota) * 100
print(f'O coeficiente de variação do tempo de resposta é de :{coef_var: .2f}%.')
```

2) Calcule média e desvio padrão da variável nota por UF.

```python
df2 = df[['Nota do Consumidor', 'UF']]
```

```python
# Média

df2.groupby('UF').mean()
```

```python
# Desvio Padrão

df2.groupby('UF').std()
```

3) Calcule a média e o desvio padrão para a variável Tempo Resposta por mês de abertura.

```python
df3 = df[['Tempo Resposta', 'Mês Abertura']] 
```

```python
# Média

df3.groupby('Mês Abertura').mean()
```

```python
# Desvio Padrão

df3.groupby('Mês Abertura').std()
```

4) Proporção de reclamações registradas e não respondidas.

```python
df4 = df[['Respondida']]

df4['Respondida'].value_counts(dropna = False, normalize = True)
```

```python
nao = df4.loc[df4['Respondida'] == "N"].count()

total = df4['Respondida'].count()

proporcao = float((nao / total) * 100)

print(f' As reclamações não respondidas representam o total de {proporcao: .2f}% do total.')
```
